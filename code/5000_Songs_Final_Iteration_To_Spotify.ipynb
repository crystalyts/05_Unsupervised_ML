```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
get_ipython().system('pip install spotipy')
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
url = "https://drive.google.com/file/d/1oYQSNxfvw6kFr6-N9rKLRAnLXlp0osEt/view?usp=share_link" 
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
Audio_5000 = pd.read_csv(path)
audioo_5000 = Audio_5000.copy()
audioo_5000.head()
```

```python
audio_5000 = audioo_5000
audio_5000.columns = audio_5000.columns.str.replace(' ', '')
audio_5000 = audio_5000.set_index('name','artist')
audio_5000 = audio_5000.drop(['id', 'html'], axis=1)
```

```python
columns = ['danceability', 'energy', 'key', 'mode', 'tempo', 'valence','loudness', 'acousticness', 'instrumentalness', 'liveness', 'duration_ms']

audio_5000_new= audio_5000.filter(columns)
audio_5000_new.head()
```

```python
audio_5000_norm = StandardScaler().fit_transform(audio_5000_new)
pd.DataFrame(audio_5000_norm, columns=columns, index=audio_5000_new.index).head()
```

```python
audio_5000_kmeans = KMeans(n_clusters=9, random_state = 123).fit(audio_5000_norm)
audio_5000_kmeans
```

```python
centroids = pd.DataFrame(audio_5000_kmeans.cluster_centers_)
centroids.columns = columns
centroids.index.name = 'cluster'
centroids.head()
```

```python
playlist_data = pd.DataFrame(audio_5000_norm, index=audio_5000_new.index)
playlist_data.columns = columns
playlist_data['cluster'] = audio_5000_kmeans.labels_ #peng_kmeans.predict(peng_norm)
playlist_data.head(20)
```

```python
playlist_data = (playlist_data
                 .merge(centroids, on='cluster', how='left')
    # .info()    
                 .assign(
                     distance = lambda x: np.sqrt(
                         np.square(x['danceability_x'] - x['danceability_y']) + 
                         np.square(x['energy_x'] - x['energy_y']) +
                         np.square(x['key_x'] - x['key_y']) +
                         np.square(x['mode_x'] - x['mode_y']) +
                         np.square(x['tempo_x'] - x['tempo_y']) +
                         np.square(x['valence_x'] - x['valence_y']) +
                         np.square(x['loudness_x'] - x['loudness_y']) +
                         np.square(x['acousticness_x'] - x['acousticness_y']) +
                         np.square(x['instrumentalness_x'] - x['instrumentalness_y']) +
                         np.square(x['liveness_x'] - x['liveness_y']) +
                         np.square(x['duration_ms_x'] - x['duration_ms_y'])
                     )   
                 )
                )
    # calulate the distance between the song and the cluster
```

```python
playlist_5000_first = playlist_data[['cluster', 'distance']]
playlist_5000_first.index.name = 'index'
playlist_5000_first
```

```python
playlist_5000_second = audioo_5000[['name', 'artist','id', 'html']]
playlist_5000_second
```

```python
playlist_5000_short = playlist_5000_second.join(playlist_5000_first)
playlist_5000_short
```

```python
playlists=(
    playlist_5000_short
    .groupby(by='cluster')
    .apply(lambda x: x.sort_values('distance'))
)
```
