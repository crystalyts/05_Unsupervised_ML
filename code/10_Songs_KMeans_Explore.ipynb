# 1. Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

# 2. Audio_10

```python
df_10 = pd.read_csv('df_audio_features_10.csv') 
df_10
```

```python
# making a copy
df_10_copy = df_10.copy()
df_10_copy
```

```python
# dropping unwanted columns
df_10_copy.drop(["html", "id", "time_signature"], axis=1, inplace=True)
df_10_copy
```

```python
# setting new indexes
df_10_copy = df_10_copy.set_index(["song_name","artist"])
df_10_copy
```

```python
# dropping another column
df_10_copy.drop("duration_ms", axis=1, inplace=True)
df_10_copy
```

```python
# dropping another column
df_10_copy.drop("mode", axis=1, inplace=True)
df_10_copy
```

# 3. Scaling the data

## Using MinMax scaler

```python
# 1. import the transformer
from sklearn.preprocessing import MinMaxScaler

# 2. initialize the transformer 
my_min_max = MinMaxScaler()

#3. Fit the transformer to the data
my_min_max.fit(df_10_copy)

#4. use the transfomer to transform the data
scaled_features = my_min_max.transform(df_10_copy)

#5. reconvert the transformed dta back to a Data Frame
scaled_features_df_10 = pd.DataFrame(scaled_features, 
                                  index=df_10_copy.index, 
                                  columns=df_10_copy.columns)

scaled_features_df_10
```

# 4. Clustering

```python
# 1. import the model
from sklearn.cluster import KMeans
 
# 2. initialize the model
my_kmeans = KMeans(n_clusters= 5, random_state = 123) # you always choose the number of k here
 
# 3. fit the model to the data
my_kmeans.fit(scaled_features_df_10) #pass your scaled data here

#4.1 obtain the cluster output (no need to use this one)
#clusters = my_kmeans.predict(scaled_features_df_10) # pass your scaled data here

# 4.2 obtain the cluster output
clusters = my_kmeans.labels_
 
# 5. attach the cluster output to our original DataFrame
scaled_features_df_10["cluster"] = clusters
```

```python
scaled_features_df_10
```

```python
# sorting results by cluster

scaled_features_df_10.sort_values(["cluster"])
```

```python
# adding names to the clusters

# 0 - Bossa Nova
# 1 - Classic
# 2 - Techno
# 3 - Rock
# 4 - Rap

cluster_names = {
    "cluster": [0, 1, 2, 3, 4],
    "cluster_name": ["Bossa Nova", "Classic", "Techno", "Rock", "Rap"]
}

cluster_names_df = pd.DataFrame(cluster_names)
```

```python
scaled_features_clusnam_df = scaled_features_df_10.merge(cluster_names_df, on="cluster")
```

```python
scaled_features_clusnam_df
```

```python
scaled_features_clusnam_df.sort_values(["cluster"])
```

# 5. Exploring the clusters

```python
scaled_features_clusnam_df.groupby(by="cluster_name").mean()
```

```python
scaled_features_clusnam_df.groupby(by="cluster_name").mean().sort_values("cluster")
```

```python
# find the coordinates of each centroid
centroids = my_kmeans.cluster_centers_

centroids_df = pd.DataFrame(centroids)
centroids_df = centroids_df.reset_index()
centroids_df = centroids_df.rename(columns = {"index": "cluster"})
centroids_df
```

```python
centroids_df = centroids_df.merge(cluster_names_df, on = "cluster")
centroids_df = centroids_df.set_index("cluster_name")
centroids_df.drop(["cluster"], axis=1, inplace=True)
centroids_df
```

```python
# calculate the euclidean distance between the centroids
from sklearn.metrics import pairwise_distances
centroid_distances = pairwise_distances(centroids_df)

sns.heatmap(centroid_distances, annot=True)

# 0 - Bossa Nova
# 1 - Classic
# 2 - Techno
# 3 - Rock
# 4 - Rap
```

# 7. Inertia / Elbow

```python
my_kmeans.inertia_
```

```python
my_kmeans.inertia_
max_k = 8

inertia_list = []

for i in range(1, max_k):
    myKMeans = KMeans(n_clusters=i, random_state = 123)
    myKMeans.fit(scaled_features_df_10)
    inertia_list.append(round(myKMeans.inertia_))
```

```python
sns.set_theme(style='darkgrid')

(
sns.relplot(x=range(1, max_k),
           y=inertia_list, 
           kind = "line",
           marker='o',
           height=8,
           aspect=2)
.set(title=f"Inertia score from a to {max_k} clusters")
.set_axis_labels("Number of clusters", "Inertia score")
);
```

# 8. Silhouette Score

```python
from sklearn.metrics import silhouette_score

silhouette_score(scaled_features_df_10, my_kmeans.labels_)
```

* negative scores are outside the cluster
* a score of 1 is perfect fit inside the cluster
* a 0 means that it is close to the boundaries

```python
max_k = 8
sil_score = []

for k in range(2, max_k):
    #prepare a kmeans with i-clusters
    myKMeans = KMeans(n_clusters=k, random_state = 123)
    #apply Kmeans on our scaled features
    myKMeans.fit(scaled_features_df_10)
    sil_score.append(silhouette_score(scaled_features_df_10, myKMeans.labels_))
```

```python
(
sns.relplot(x=range(2, max_k),
           y=sil_score, 
           kind = "line",
           marker='o',
           height=8,
           aspect=2)
.set(title=f"Silhouette score from 2 to {max_k} clusters")
.set_axis_labels("Number of clusters", "Silhouette score")
);
```
